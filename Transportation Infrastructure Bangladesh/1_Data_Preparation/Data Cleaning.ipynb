### Fill the  NaNs and Zero by LRP

```python
# Create a series of all LRPs from the NaN and value 0 coordinates

# First, create series with LRPs of NaNs (94)
LRP = file[file['lon'].isnull()==True]['LRPName']

# Then, append LRPs from zeros to that series (17)
LRP = LRP.append(file[file['lat']==0.0]['LRPName'])

# Check, should be 94+17=111
len(LRP)
```

```python
# There are only 76 unique LRPs that have no coordinates
len(LRP.unique())
```

```python
# Create a dictionary with the LRPs to be found
# if the LRP is already in the dictionary, skip it 
# for now, the value is set to 1

d = {}

for x in LRP:
    if x in d:
        continue
    else: 
        d[x]=1

# Check the length of the dictionary
len(d)
```

```python
# Get the right coordinates from other rows in the dataframe with the same LRP
# creates a dictionary with LRPS and coordinates that can be used to fill the NaNs and zeros
for key in d: 
    for index, row in file.iterrows():
        if row['LRPName'] == key:
            d[key] = (row['lat'],row['lon'])
            break
```

```python
# Check what dictionary looks like; only one NaN and one Zero left
d
```

```python
# Create function to spot the NaNs 
def isNaN(num):
    return num != num

# Fill the NaN's and zero's with the dictionary 
for index, row in file.iterrows():
    if row['lon'] == 0.0:
        for key in d:
            if key == row['LRPName']:
                file.loc[index, 'lon'] = d[key][1]
                file.loc[index, 'lat'] = d[key][0]
                print ('zeros gone!')
    if isNaN (row['lon']):
        for key in d:
            if key == row['LRPName']:
                file.loc[index, 'lon'] = d[key][1]
                file.loc[index, 'lat'] = d[key][0]
                print ('nans gone!')
```

```python
# Check the remaining zeros and nans (both 1)
len(file[file['lat']==0.0])
```

### Switching coordinates

```python
# When exploring in Java, a select group (15) of bridges had coordinates in the wrong column 

# A dataframe is made containing all the bridges that have their lat and lon in the wrong column 
df_sw = file[(file['lon'] > 20) & (file['lon'] < 27) & (file['lat'] > 88) & (file['lat'] < 93)]

df_sw.head(5)
```

```python
# An iteration over all the columns in that dataframe and switch the corresponding lat and lon in the original dataframe

for index, row in df_sw.iterrows():
    file.loc[index, 'lat'] = df_sw.loc[index, 'lon']
    file.loc[index, 'lon'] = df_sw.loc[index, 'lat']
```

```python
# Again, a dataframe is made from the original dataframe containing all rows with incorrect coordinates (should now be zero)

check = file[(file['lon'] > 20) & (file['lon'] < 27) & (file['lat'] > 88) & (file['lat'] < 93)]
check.head()
```

### Check whether there are coordinates far outside of Bangladesh

```python
# A dataframe is made containing all the bridges that are outside of Bangladesh
# There are 22 rows

outside = file[(file['lat'] < 20.747) | (file['lat'] > 26.631) | (file['lon'] < 88) | (file['lon'] > 92.686)]
outside
```

```python
# There are 19 unique LRPs to be found to use to fix the coordinates

LRP_outside = outside['LRPName']
len(LRP_outside.unique())
```

```python
# Create a dictionary with the LRPs to be found
# if the LRP is already in the dictionary, skip it 
# for now, the value is set to 1

d = {}

for x in LRP_outside:
    if x in d:
        continue
    else: 
        d[x]=1

# Check the length of the dictionary
len(d)
```

```python
# Get the right coordinates from other rows in the dataframe with the same LRP
# Create a dictionary with LRPS and coordinates that can be used to fill the NaNs and zeros
for key in d: 
    for index, row in file.iterrows():
        if row['LRPName'] == key:
            d[key] = (row['lat'],row['lon'])
            break
```

```python
# Fill the coordinates that lay outside of the country with the dictionary 

for a, b in outside.iterrows():
    for index, row in file.iterrows():
        if b['lon'] == row['lon']:
            print('Change this bridges coordinates!')
            for key in d:
                if key == row['LRPName']:
                    file.loc[index, 'lon'] = d[key][1]
                    file.loc[index, 'lat'] = d[key][0]
```

```python
# All bridges but one of the 22 are fixed! 

outsidecheck = file[(file['lat'] < 20.747) | (file['lat'] > 26.631) | (file['lon'] < 88) | (file['lon'] > 92.686)]
outsidecheck
```

### General check for big differences in coordinates of bridges on the same roads

```python
# Create empty dataframes with columns of original df to fill with bridges 
# that are too far off the others based on the lat or lon

columns = list(file.columns)

df_ = pd.DataFrame(columns=columns)
```

```python
# For loop that checks per row whether there is are difference bigger than one point between either the longitude or latitude 
# of the succeeding bridge, IF they are on the same road
# Produces a dataframe containing bridges with incorrect longitudes 

for index, row in file.iterrows():
    nextrow = index+1
    

    if index == file.index[-1]: # Stop at last line since there exists no row+1
        break

    elif file.iloc[index+1, 0]==row['road']:        # Checks whether the row and row+1 are on the same road
        if (row['lat']- file.iloc[index+1, 17])>1:  # If so, check whether the lats/lons have a difference bigger than one point
            #print (row['lat']- file.iloc[index+1, 17])
            #print ('too big of a difference!')
            #print (row['name'])
            df_ = df_.append(row)
        if ( row['lon']-file.iloc[index+1, 18])>1: #If so, check whether the longitudes have a difference bigger than one point
            #print ( row['lon']-file.iloc[index+1, 18])
            #print ('too big of a difference!')
            #print (row['name'])
            df_ = df_.append(row)
    else: 
        continue
```

```python
# There are 125 bridges with coordinates with one point difference to there successor of the same road
len(df_)
```

```python
# There are 83 unique LRPs to be found to use to fix the coordinates

LRP_difference = df_['LRPName']
len(LRP_difference.unique())
```

```python
# Create a dictionary with the LRPs to be found
# if the LRP is already in the dictionary, skip it 
# for now, the value is set to 1

d = {}

for x in LRP_difference:
    if x in d:
        continue
    else: 
        d[x]=1

# Check the length of the dictionary
len(d)
```

```python
# Get the right coordinates from other rows in the dataframe with the same LRP
# Create a dictionary with LRPS and coordinates that can be used to changes the wrong coordinates

for key in d: 
    for index, row in file.iterrows():
        if row['LRPName'] == key:
            d[key] = (row['lat'],row['lon'])
            break
```

```python
# Fill the coordinates that are wrong with the dictionary 

for a, b in df_.iterrows():
    for index, row in file.iterrows():
        if b['lon'] == row['lon']:
            for key in d:
                if key == row['LRPName']:
                    file.loc[index, 'lon'] = d[key][1]
                    file.loc[index, 'lat'] = d[key][0]
```

```python
# CHECK shows that only 125 - 111 = 14 bridges could be fixed with other LRP coordinates. 
# it can be concluded that 111 bridges have a big difference due to wrong LRP estimations. 

columns = list(file.columns)
check = pd.DataFrame(columns=columns)

for index, row in file.iterrows():
    nextrow = index+1
    if index == file.index[-1]: # Stop at last line since there exists no row+1
        break
    elif file.iloc[index+1, 0]==row['road']:        # Checks whether the row and row+1 are on the same road
        if (row['lat']- file.iloc[index+1, 17])>1:  # If so, check whether the lats/lons have a difference bigger than one point
            #print (row['lat']- file.iloc[index+1, 17])
            #print ('too big of a difference!')
            #print (row['name'])
            check = check.append(row)
        if ( row['lon']-file.iloc[index+1, 18])>1: #If so, check whether the longitudes have a difference bigger than one point
            #print ( row['lon']-file.iloc[index+1, 18])
            #print ('too big of a difference!')
            #print (row['name'])
            check = check.append(row)
    else: 
        continue
check
```
